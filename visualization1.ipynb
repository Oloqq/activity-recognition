```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

PATH = 'data/unpacked/wchodzenie-2023-10-18_21-37-04/'
unpacked = 'data/unpacked/'

df = pd.read_csv(f'{PATH}Accelerometer.csv')
df.head()
```

# Example 

```python

# 500 samples 
start = 1000
end = 1500

X_time, x, y, z = df['seconds_elapsed'][start:end], df['x'][start:end], df['y'][start:end], df['z'][start:end]

```

```python
# Walking up the stairs - all axis on seperate plots

fig, axs = plt.subplots(3, 1)
plt.xlabel("time [s]")

axs[0].plot(X_time, x)
axs[0].set_title('x [m/s2]')

axs[1].plot(X_time, y)
axs[1].set_title('y [m/s2]')

axs[2].plot(X_time, z)
axs[2].set_title('z [m/s2]')

fig.set_size_inches(20 ,15)
plt.show()
```

```python
# Walking up the stairs - all axis on one plot

fig = plt.gcf()
fig.set_size_inches(15, 8)
plt.title("Accelerometer - walking up stairs")

plt.plot(X_time, x, label="x")
plt.plot(X_time, y, label="y")
plt.plot(X_time, z, label="z")

plt.legend()
plt.show()
```

# Accelerometer data
Comparizon of the accelerometer data for different activities. 

```python
def showAccPlotsForFiles(dirs):
    for dir in dirs:
        df = pd.read_csv(f'{unpacked}{dir}/Accelerometer.csv')
        df.head()

        start = 1000
        end = 1600

        X_time, x, y, z = df['seconds_elapsed'][start:end], df['x'][start:end], df['y'][start:end], df['z'][start:end]

        fig = plt.gcf()
        fig.set_size_inches(15, 8)
        plt.title(f"Accelerometer - {dir}")

        plt.plot(X_time, x, label="x")
        plt.plot(X_time, y, label="y")
        plt.plot(X_time, z, label="z")

        plt.ylabel("[m/s^2]")
        plt.xlabel("time [s]")

        plt.xlim(10, 15)
        plt.ylim(-15, 15)

        plt.legend()
        plt.show()
```

```python
examples = [
    'wchodzenie-2023-10-18_21-37-04', 
    'schodzenie-2023-10-18_14-14-49', 
    'chodzenie-2023-10-18_21-24-21', 
    'chodzenie_reka-2023-10-18_17-14-22'
]

showAccPlotsForFiles(examples)
```

# Orientation data - roll pitch yaw


```python
def showOrienPlotsForFiles(dirs):
    for dir in dirs:
        df = pd.read_csv(f'{unpacked}{dir}/Orientation.csv')
        df.head()

        start = 1500
        end = 2000

        X_time, qz, qy, qx, qw = df['seconds_elapsed'][start:end], df['qz'][start:end], df['qy'][start:end], df['qx'][start:end], df['qw'][start:end]

        fig, axs = plt.subplots(2, 1)

        fig = plt.gcf()
        fig.set_size_inches(15, 8)

        axs[0].plot(X_time, qx, label="qx")
        axs[0].plot(X_time, qy, label="qy")
        axs[0].plot(X_time, qz, label="qz")
        axs[0].plot(X_time, qw, label="qw")
        axs[0].legend()
        axs[0].title.set_text(f"Orinetation - quaternions - {dir}")

        # axs[0].ylabel("[rad/s]")
        # axs[0].xlabel("time [s]")

        roll, pitch, yaw = df['roll'][start:end], df['pitch'][start:end], df['yaw'][start:end]

        axs[1].plot(X_time, roll, label="roll")
        axs[1].plot(X_time, pitch, label="pitch")
        axs[1].plot(X_time, yaw, label="yaw")
        axs[1].legend()
        axs[1].title.set_text(f"Orinetation - pitch roll yaw - {dir}")

        plt.show()
```

```python
showOrienPlotsForFiles(examples)
```
